```python
import torch.utils.data as td
import torchvision.utils as tvu
import data_sets as ds
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
```

```python
SynDataSet = ds.SynData("syn_data/syn_input", "syn_data/syn_label")
```

```python
train, validate, test = td.random_split(SynDataSet, [0.6,0.2,0.2])
my_data = [train,validate,test]
```

```python
for subset in my_data:
    print("Split Here")
    SubsetLoader = td.DataLoader(subset, batch_size=5)
    SubsetIterator = iter(SubsetLoader)
    for idx in range(len(SubsetIterator)):
        image, label = next(SubsetIterator)
        grid = tvu.make_grid(image) #.numpy()[0] hack to show tensor in plt
        plt.imshow(grid.numpy()[0], cmap="gray", vmin=0, vmax=255)
        plt.show()
        #lab_grid = tvu.make_grid(label)
        #plt.imshow(lab_grid.numpy()[0], cmap="gray", vmin=0, vmax=1)
        #plt.show()
```
